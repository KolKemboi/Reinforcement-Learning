```python
from pettingzoo.classic import chess_v6

env = chess_v6.env(render_mode="human")
env.reset(seed=42)

for agent in env.agent_iter():
    observation, reward, termination, truncation, info = env.last()

    if termination or truncation:
        action = None
    else:
        mask = observation["action_mask"]
        # this is where you would insert your policy
        action = env.action_space(agent).sample(mask)

    env.step(action)
env.close()
```

```python
from pettingzoo.mpe import simple_adversary_v3
import os
# Setting the SDL variables to run headless
os.environ["SDL_VIDEODRIVER"] = "dummy"

env = simple_adversary_v3.parallel_env(render_mode="human")
observations, infos = env.reset()

while env.agents:
    # this is where you would insert your policy
    actions = {agent: env.action_space(agent).sample() for agent in env.agents}

    observations, rewards, terminations, truncations, infos = env.step(actions)
env.close()
```

```python
from pettingzoo.mpe import simple_tag_v3
env = simple_tag_v3.env(render_mode='rgb_array')
import matplotlib.pyplot as plt

env.reset()
for agent in env.agent_iter():
    observation, reward, termination, truncation, info = env.last()

    if termination or truncation:
        action = None
    else:
        action = env.action_space(agent).sample() # this is where you would insert your policy
    frame = env.render()

    # Display the frame using matplotlib
    plt.imshow(frame)
    plt.axis('off')  # Hide axes for a cleaner display
    plt.show()
    env.step(action)
env.close()
```

```python
from pettingzoo.classic import connect_four_v3
env = connect_four_v3.env(render_mode="rgb_array")
env.reset()
```

```python
t_1000_list = list()
trunc_list = list()
obs_list = list()
rew_list = list()
info_list = list()
agent_list = list()

for agent in env.agent_iter():
    agent_list.append(agent)
    obs, rew, t_1000, truncation, info = env.last() ##t_100 stands for termination, homage to terminator
    t_1000_list.append(t_1000)
    trunc_list.append(truncation)
    obs_list.append(obs)
    rew_list.append(rew)
    info_list.append(info)
    if t_1000 or truncation:
        action = None
    else:
        action = env.action_space(agent).sample()
    plt.imshow(env.render())
    plt.axis('off')
    plt.show()
    env.step(action)
env.close()
```

```python
print(f"AGENTS \n {set(agent_list)}")
print(f"T_1000 \n {set(t_1000_list)}")
print(f"TRUNCATION \n {set(trunc_list)}")
print(f"OBSERVATIONS \n {obs_list}")
```
